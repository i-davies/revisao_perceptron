```python
# Entradas
renda_mensal = 2.0 # 0 a 10 representando faixas salarias
historico_pagador = 10.0 # 0 a 10 bom pagador
```

```python
# Pesos
peso_renda = 0.3
peso_historico = 0.8

# Bias (A Resistência natural em empresas dinheiro)
bias = -7.0
```

```python
# Calculo Z (Equação de Decisão)
z = (renda_mensal * peso_renda) + (historico_pagador * peso_historico) + bias

print(f"O valor interno gerado pelo Perceptron = {z:.2f}")
```

```python
# Função de Ativação (Degrau / Step)
if z >= 0:
    decisao = 1
    motivo = "Aprovado! Bom pagador."
else:
    decisao = 0
    motivo = "Reprovado! Risco muito alto."

print(f"Decisão Final {decisao}")
print(f"Motivo {motivo}")
```
